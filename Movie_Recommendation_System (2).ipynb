```python
import pandas as pd
import numpy as np
import ast
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv') #đọc dữ liệu
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head() #xem qua data
```

```python
movies.head() 
```

```python
movies = movies.merge(credits, left_on='title', right_on='title') #merge 2 file data và 1 file movie
```

```python
movies.head() #check lại data
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']] #chỉ giữ lại những cột này
```

```python
movies.head(1) #check lại data
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
#Hàm convert(obj): Chuyển một chuỗi biểu diễn danh sách các đối tượng (dict) có khóa 'name' thành một danh sách chỉ chứa các giá trị 'name'.
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
movies['cast'] = movies['cast'].apply(lambda x: [i['name'] for i in ast.literal_eval(x)[:3]])  
# Chuyển chuỗi JSON đó thành danh sách thật của các dict Python
# Only top 3 actors
# Tạo danh sách chỉ chứa tên diễn viên
```

```python
movies['crew'] = movies['crew'].apply(lambda x: [i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director'])
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']

# tag của movie là gồm các cột genres, keywwords, cast, crew
```

```python
movies['tags']
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x)) #join lại bằng dấu cách
```

```python
movies['tags']
```

```python
movies = movies[['movie_id', 'title', 'overview', 'tags']] #giờ dataframe movies có cột movies tag sẽ là tổng hợp của genres, keywords, cast, crew luôn
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower()) 
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['tags'])

# chuyển vb thành vector số học theo thuật toán td idf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# tính độ tương đồng cosine giữa các phim
# kết quả là ma trận vuông nxn (n là số phim)
# cosine_sim[i][j] độ giống nhau giữa phim i và phim j, range từ 0->1, 0 là hoàn toàn klq, 1 là giống
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movies[movies['title'] == title].index[0] # lấy index của phim có title = title cần tìm
    sim_scores = list(enumerate(cosine_sim[idx])) # cosine_sim[index] để lấy độ tương đồng của phim idx với các phim khác. VD cosine_sim[0] = [1.0, 0.43, 0.12, 0.09, ...]
    #enumurate gắn chỉ số phim tương ứng với giá trị similarity. VD [(0, 1.0), (1, 0.43), (2, 0.12), (3, 0.09), ...]
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True) #sort theo độ similarity
    sim_scores = sim_scores[1:11]  # Get top 10 similar movies
    movie_indices = [i[0] for i in sim_scores] # lấy ds chỉ số của 10 phim đó
    return movies['title'].iloc[movie_indices] # trả về title 10 phim
```

```python
print(get_recommendations('The Dark Knight Rises'))
```
